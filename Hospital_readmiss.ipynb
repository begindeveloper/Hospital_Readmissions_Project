# data cleaning and exploration process

```python
# find the missing values
# return true if there is a NaN
df_missing_values = df.isnull()
print(df_missing_values)
```

```python
# Missing values per column
df.isna().sum()
```

```python
df_cleaned = df.dropna()
print(df_cleaned)
```

```python
df_duplicate = df_cleaned.drop_duplicates()
print(df_duplicate)
```

```python
# covert columns to their correct data types
columns_to_scale = ['Score', 'Lower Estimate', 'Higher Estimate', 'Number of Patients', 'Number of Patients Returned']
```

```python
# First, replace the problematic 'Not Available' strings with NaN
df.replace('Not Available', np.nan, inplace=True)
print("Replaced 'Not Available' strings with NaN.")

for column in columns_to_scale:
    df[column] = pd.to_numeric(df[column], errors='coerce')
    print(f"converted '{column}' to numerical data type.")
```

```python
# Now, we will fill the missing numerical values with the median.
# This is a good practice as it is not influenced by outliers.
for column in columns_to_scale:
    if column in df.columns and df[column].isnull().sum() > 0:
        median_value = df[column].median()
        df[column].fillna(median_value)
        print(f"Filled {column} with its median value: {median_value}")
```

```python
#categorical columns with many missing values
columns_to_impute = ['Footnote']

for col in columns_to_impute:
    df[col].fillna('Unknown')
    print("Filled missing 'Footnote' values with Unknown")
```

```python
# Convert 'Start Date' and 'End Date' to datetime objects
if 'Start Date' in df.columns:
    df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
    print("Converted 'Start Date' to datetime format.")

if 'End Date' in df.columns:
    df['End Date'] = pd.to_datetime(df['End Date'], errors='coerce')
    print("Converted 'End Date' to datetime format.")
```

```python
# Calculate the length of stay and store it in a new column
df['length_of_stay'] = df['End Date'] - df['Start Date'] 

# convert to integer number of days
df['length_of_stay'] = df['length_of_stay'].dt.days

# print out the results
print(df[['Start Date', 'End Date', 'length_of_stay']].head())
```

```python
df_missing_values_after = df.isnull().sum()
print(df_missing_values_after)
```

# EXPLORATORY DATA ANALYSIS (EDA)

```python
# What are the most common diagnoses or procedures associated with readmission?
if 'Measure Name' in df.columns:
    top_measures = df['Measure Name'].value_counts().nlargest(10)
    print("\nTop 10 Most Common Measures/Diagnoses:")
    print(top_measures)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(x=top_measures.values, y=top_measures.index)
    plt.title('Top 10 Most Common Unplanned Visit Measures')
    plt.xlabel('Number of Visits')
    plt.ylabel('Measure Name')
    plt.show()
```

```python
# Is there a specific length of hospital stay that correlates with higher or lower readmission rates?
plt.figure(figsize=(10,6))
sns.histplot(df['length_of_stay'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Unplanned Visit Length of Stay (Calculated)')
plt.xlabel('Length of Stay (Days)')
plt.ylabel('Frequency')
plt.show()
```

```python
# How do unplanned visits vary over time?
if 'End Date' in df.columns:
    # We will now group by month to see a more granular trend over time
    df['End Month'] = df['End Date'].dt.strftime('%Y-%m')
    visits_by_month = df.groupby('End Month').size()
    
    print("\nTotal visits by month:")
    print(visits_by_month)
    
    plt.figure(figsize=(12, 6))
    visits_by_month.plot(kind='line', marker='o', color='purple')
    plt.title('Unplanned Visits Over Time (by Month)')
    plt.xlabel('Month')
    plt.ylabel('Number of Visits')
    plt.grid(True)
    plt.show()
```
